```python
def read(path):
    stacks = []
    
    linea = 0
    qStacks = 0
    h = 0
    #Lee el archivo
    with open(path) as file:
        for line in file:
            s = line.split()
            
            #STACKS
            if linea == 1:
                qStacks = s[2]
                #print(qStacks)
            #ALTURA
            if linea == 2:
                h = s[2]
                #print(h)
            
            if linea == 3:
                contenedores = s[2]
            #COLUMNAS
            if linea > 3 :
                temporalStack = []
                #Size of stack
                sos = len(s)-3
                
                #Ahora se guarda en el stack temporal
                temporalStack.append(sos)
                temporalStack.append(" ")
                for i in range(0,len(s)):
                    #Agrega el numero y un " "
                    if i > 2:
                        temporalStack.append(s[i])
                        if i < len(s)-1:
                            temporalStack.append(" ")
                        
                #print(temporalStack)
                stacks.append(temporalStack)
            #print(s)
            linea+=1

    return qStacks,contenedores,stacks

read("Referencias\BF1\cpmp_16_5_48_10_29_1.bay")
```
